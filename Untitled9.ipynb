```python
import pandas as pd

df = pd.read_csv(
    "/content/sample_data/caracteristicasautista.csv",
    sep=";",
    engine="python",
    encoding="latin1"
)

df
```

```python
# Exemplo de lista
sintomas = ["atrasos na fala", "atraso na fala", "comportamentos repetitivos", "comportamento repetitivo"]

# passando nomes para o singular
def para_singular(texto):
    texto = texto.strip().lower()
    if texto.endswith("ções"):  # ex.: interações -> interação
        texto = texto[:-3] + "ão"
    elif texto.endswith("s") and not texto.endswith("as") and not texto.endswith("os"):
        texto = texto[:-1]  # remove 's' final se for plural comum
    elif texto.endswith("os") or texto.endswith("as"):
        texto = texto[:-2]  # remove plural comum masculino/feminino
    return texto

# Aplica
sintomas_sing = [para_singular(t) for t in sintomas]

print(sintomas_sing)
```

```python
dados_expandidos = []

# Quebra os sintomas por vírgula e cria novas linhas
for _, linha in df.iterrows():
    nivel = linha["nivel_suporte"].strip().lower()
    sintomas = str(linha["sintomas_observados"]).split(",")
    for sintoma in sintomas:
        sintoma_limpo = sintoma.strip().lower()
        dados_expandidos.append([nivel, sintoma_limpo])

# Diagrama com Sintomas separados
df_exp = pd.DataFrame(dados_expandidos, columns=["nivel_suporte", "sintoma"])

# frequencia dos niveis de sintomas
contagem = df_exp.groupby(["nivel_suporte", "sintoma"]).size().reset_index(name="frequencia")

# Ordem de aparecimento de caracteritsicas por niveis
contagem = contagem.sort_values(by=["nivel_suporte", "frequencia"], ascending=[True, False])

# Mostrar o resultado
print("=== Todas as frequências por nível ===")
print(contagem)

# Top 10 de cada nível
top5 = contagem.groupby("nivel_suporte").head(10)
print("\n=== Top 5 sintomas mais comuns por nível de suporte ===")
print(top5)
```
